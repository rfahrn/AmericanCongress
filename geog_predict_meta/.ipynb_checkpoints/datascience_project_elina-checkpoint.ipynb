```python
columns = ["speakerid","speech_id","lastname","firstname","chamber","state","gender","party","district","nonvoting"]
# make a data frame of all the files
with open("DS_projet/Us hein daily speaker map/097_SpeakerMap.txt", newline='') as speaker_map_97:
    # split at "|"
    spamreader_97 = csv.reader(speaker_map_97, delimiter="|")
    # set column names
    df_97 = pd.DataFrame(spamreader_97, columns = columns)
    # delete header row
    df_97 = df_97[1:]
#df_97.head(10)
```
